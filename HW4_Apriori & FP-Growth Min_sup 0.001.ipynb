```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn import datasets
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from apyori import apriori as asp_apr
```

```python
df1 = pd.read_excel("Trade dataset_Pre Over.xlsx",header = None)
df1
```

```python
records = []
for i in range(0,36498):
    records.append([str(df1.values[i,j]) for j in range(0,251)])
```

```python
te = TransactionEncoder()
te_ary  = te.fit(records).transform(records)
te_ary
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df = df.drop(labels=['nan'],axis=1)
df
```

## Apriori

```python
start_apr = time.time()
frequent_itemsets_apr = apriori(df, min_support = 0.001, use_colnames = True, low_memory= True)
frequent_itemsets_apr
```

```python
frequent_itemsets_apr.sort_values(by= "support", ascending= False).head(20)
```

```python
rules_apr = association_rules(frequent_itemsets_apr, metric = "confidence", min_threshold = 0.8)
end_apr = time.time()
rules_apr
```

```python
def arl_recommender(rules_df, product_id, recommendation_number=1):
    recommendation_list_apr = []
    sorted_rules_apr = rules_apr.sort_values("lift", ascending=False)
    for idx, product in enumerate(sorted_rules_apr["antecedents"]):
        for j in list(product):
            if j == product_id:
              recommendation_list_apr.append(list(sorted_rules_apr.iloc[idx]["consequents"])[0])
    return set(recommendation_list_apr[:recommendation_number])
```

```python
arl_recommender(rules_apr, "BAV99,215", 30)
```

## FP Growth

```python
fpgrowth(df, min_support = 0.001)
```

```python
start_fp = time.time()
frequent_itemsets_fp = fpgrowth(df, min_support = 0.001, use_colnames = True)
frequent_itemsets_fp
```

```python
frequent_itemsets_fp.sort_values(by= "support", ascending= False).head(20)
```

```python
rules_fp = association_rules(frequent_itemsets_fp, metric = "confidence", min_threshold = 0.8)
end_fp = time.time()
rules_fp
```

```python
def arl_recommender_fp(rules_df, product_id, recommendation_number=1):
    recommendation_list_fp = []
    sorted_rules_fp = rules_fp.sort_values("lift", ascending=False)
    for idx, product in enumerate(sorted_rules_fp["antecedents"]):
        for j in list(product):
            if j == product_id:
              recommendation_list_fp.append(list(sorted_rules_fp.iloc[idx]["consequents"])[0])
    return set(recommendation_list_fp[:recommendation_number])
```

```python
arl_recommender_fp(rules_fp, "SR2351C", 30)
```

```python
print("Apriori 所花費時間:", end_apr - start_apr)
print("FP-Growth 所花費時間:", end_fp - start_fp)
```
